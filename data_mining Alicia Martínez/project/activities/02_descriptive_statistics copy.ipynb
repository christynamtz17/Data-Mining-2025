```python
import pandas as pd
from utils import path_loader as p
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Load csv file
try:
    df = pd.read_csv(p.generate_file_path(
        file_name='avocado_prices_processed', file_extension='csv', directory='output'))

    n_row, n_col = df.shape
    print(f'CSV File loaded succesfully!\nThere are {n_row} rows and {n_col} columns')
except Exception as e:
    print(f'Something wrong happended!: {e}')
```

```python
# Display descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
# Identify entities and relations
# Assuming 'entities' are categorical columns and relations are connections or numerical correlations
entities = df.select_dtypes(include=['object', 'category']).columns
relations = df.select_dtypes(include=['float64', 'int64']).columns
print(f"\nIdentified Entities: {list(entities)}")
print(f"Identified Relations (Numeric Columns): {list(relations)}")

# Group by entities and calculate statistics
grouped_stats = {}
for entity in entities:
    grouped_stats[entity] = df.groupby(entity).describe()

# Display grouped statistics
for entity, stats in grouped_stats.items():
    print(f"\nGrouped Statistics for '{entity}':")
    print(stats)
```
